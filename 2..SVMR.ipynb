```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset=dataset.astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
 from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)#assign splitted test and train dataset to diff variables
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="poly",C=1000,coef0=0.1)
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
regressor.intercept_
```

```python
from sklearn.metrics import r2_score
```

```python
regressor.n_support_
```

```python
r_score=r2_score(y_test,y_pred)#passing arguments as predited and actual output r2 function for Evaluatation
r_score
```
