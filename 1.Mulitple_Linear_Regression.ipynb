```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)         
```

```python
dataset=dataset.astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
 from sklearn.model_selection import train_test_split
```

```python
train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)#assign splitted test and train dataset to diff variables
```

```python
x_train
```

```python
#model creation
from sklearn.linear_model import LinearRegression#import linearRegression class
regressor=LinearRegression()#assign class to variable
regressor.fit(x_train,y_train)
```

```python
weight=regressor.coef_
weight
```

```python
#display the bias
bias=regressor.intercept_
bias
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score#Evaluate the model using R2
```

```python
r_score=r2_score(y_test,y_pred)#passing arguments as predited and actual output r2 function for Evaluatation
r_score#display r2 value
```
